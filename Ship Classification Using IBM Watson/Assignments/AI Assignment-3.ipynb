```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(rescale=(1./255),featurewise_center=True,horizontal_flip=True,vertical_flip=True,shear_range=0.2,featurewise_std_normalization=True)
test_gen = ImageDataGenerator(rescale=(1./255))
```

```python
train = train_gen.flow_from_directory(r'Users\koushalvel\Desktop\Bird Species\train_data\train_data',
                                      target_size=(120,120),
                                      class_mode='categorical',
                                      batch_size=16)
test = test_gen.flow_from_directory(r'Users\koushalvel\Desktop\Bird Species\test_data\test_data',
                                    target_size=(120,120),
                                    class_mode='categorical',
                                    batch_size=16)
```

## CNN

```python
from keras.layers import Convolution2D, MaxPooling2D, Flatten,Dense
from keras.models import Sequential
from keras.layers import BatchNormalization, Dropout
from keras.regularizers import l2
```

```python
model = Sequential()

model.add(Convolution2D(100,(3,3),activation='relu',input_shape=(120,120,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Convolution2D(150,(3,3),activation='relu',dilation_rate=2))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Convolution2D(180,(3,3),activation='relu',dilation_rate=2))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Convolution2D(180,(3,3),activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Convolution2D(95,(3,3),activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Flatten())

model.add(Dense(105,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(140,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(145,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(100,activation='relu'))

model.add(Dense(16,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='accuracy',patience=10)
```

```python
model.fit(train,batch_size=1,validation_data=test,epochs=1000,callbacks=early_stop)
```

```python
model.save('model.h5')
```
